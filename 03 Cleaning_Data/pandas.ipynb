```python
#identifying null values variables

isnull=df.isnull()
isnull
```

```python
isnull_count= isnull.sum()
print(f"total number of null values {isnull_count}")
```

```python
#total nulls in dataset
total_isnull_count=isnull_count.sum().sum()
print(f"total  nulls {total_isnull_count}")
```

```python
# replace single quote with blank and make the date formatted

df['Date'] = df['Date'].replace("'","",regex=True)
df
```

```python
#Handle null values, return dataframe with no empty cells
#we can use df.dropna(), it simply removes empty cells but will not change the dataframe,
#if we want the dataframe to be changed, simply use inplace=True

new_df=df.dropna()
new_df
```

```python
isnull_count_newdf= new_df.isnull().sum().sum()
isnull_count_newdf
```

```python
# another way of handling the null values is by replacing the null values
df.fillna(130,inplace=True)
df
```
